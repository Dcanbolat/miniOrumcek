XML Oluşturma ve Yazma

```python
import xml.etree.ElementTree as ET
import ipywidgets as widgets
from IPython.display import display

# Form alanlarını oluşturuyoruz
title_widget = widgets.Text(description="Başlık:")
subject_widget = widgets.Text(description="Anahtar Kelimeler:")
description_widget = widgets.Text(description="Özet bilgi:")
creator_widget = widgets.Text(description="Oluşturan kişi / kuruluş:")
license_widget = widgets.Text(description="Açık Lisans:")
language_widget = widgets.Text(description="Dil:")
identifier_widget = widgets.Text(description="Kaynağın URL'si:")

# Formun oluşturulması ve yayınlanması
form = widgets.VBox([title_widget, subject_widget, description_widget, creator_widget, license_widget, language_widget, identifier_widget])
display(form)

# Verileri XML dosyasına ekleyen fonksiyon
def add_data_to_xml(b):
    # Formdan gelen verilerin alınması
    title = title_widget.value
    subject = subject_widget.value
    description = description_widget.value
    creator = creator_widget.value
    license = license_widget.value
    language = language_widget.value
    identifier = identifier_widget.value

    # XML dosyasının oluşturulması veya var olanın açılması
    try:
        tree = ET.parse('metadata.xml')
        root = tree.getroot()
    except FileNotFoundError:
        root = ET.Element('OERs')
        tree = ET.ElementTree(root)

    # Yeni bir kaynak elementi oluşturup verileri ekleme
    kaynak = ET.SubElement(root, 'OER')
    ET.SubElement(kaynak, 'Title').text = title
    ET.SubElement(kaynak, 'Subject').text = subject
    ET.SubElement(kaynak, 'Description').text = description
    ET.SubElement(kaynak, 'Creator').text = creator
    ET.SubElement(kaynak, 'License').text = license
    ET.SubElement(kaynak, 'Language').text = language
    ET.SubElement(kaynak, 'Identifier').text = identifier

    # Verilerin XML dosyasına kaydedilmesi
    tree.write('metadata.xml', encoding='utf-8', xml_declaration=True)

    print("Veriler XML dosyasına eklendi")

# Kaydet butonunu oluşturma
button = widgets.Button(description="Kaydet")
display(button)
```
